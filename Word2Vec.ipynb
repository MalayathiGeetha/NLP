```python
#CBOW AND SkipGram

from gensim.models import Word2Vec

sentences = [
    ['i', 'love', 'machine', 'learning'],
    ['deep', 'learning', 'is', 'fun'],
    ['natural', 'language', 'processing', 'is', 'interesting'],
    ['i', 'love', 'deep', 'learning']
]
cbow_model = Word2Vec(
    sentences,
    vector_size=50,
    window=2,
    min_count=1,
    sg=0,             # 0 for CBOW, 1 for Skip-gram
    epochs=100
)
print("Vector for 'learning':\n", cbow_model.wv['learning'])

print("Most similar to 'learning':", cbow_model.wv.most_similar('learning'))
```

```python
# Skip-gram model: sg=1
skipgram_model = Word2Vec(
    sentences,
    vector_size=50,
    window=2,
    min_count=1,
    sg=1,        # 1 for Skip-gram
    epochs=100
)

# Check vector for a word
print("Vector for 'learning':\n", skipgram_model.wv['learning'])

# Most similar words
print("Most similar to 'learning':", skipgram_model.wv.most_similar('learning'))
```
